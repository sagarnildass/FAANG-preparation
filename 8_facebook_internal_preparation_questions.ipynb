# 1. Passing Yearbooks


```python
def findSignatureCounts(arr):
  # Write your code here
    hash_map = {}
    result = []

    n = len(arr)

    for i in range(1, n+1):
        hash_map[i] = 1


    for i in range(n):
        if arr[i] != i+1:
            hash_map[arr[i]] += 1

    for i in hash_map.keys():
        result.append(hash_map[i])
    return result
```

```python
findSignatureCounts([2,1])
```

```python
findSignatureCounts([1,2])
```

```python
findSignatureCounts([3,1,2])
```

# 2. Rotational Cipher

```python
def rotationalCipher(input_str, rotation_factor):
    # Write your code here
    
    cap_alpha = [chr(c) for c in range(ord('A'), ord('Z') + 1)]
    alpha  = [chr(c) for c in range(ord('a'), ord('z') + 1)]
    
    result = []
    
    for c in input_str:
        if not c.isalnum():
            result.append(c)
        elif c.isalpha():
            if c.isupper():
                idx = cap_alpha.index(c) + rotation_factor
                if idx > 25:
                    idx %= 26
                result.append(cap_alpha[idx])
            else:
                idx = alpha.index(c) + rotation_factor
                if idx > 25:
                    idx %= 26
                result.append(alpha[idx])
        elif c.isdigit():
            result.append(str(int(c) + rotation_factor)[-1])
    return "".join(result)
```

# 3. Pair Sums

```python
def numberOfWays(arr, k):
    c = 0
    while arr:
      s = arr.pop(0)
      for i in range(0,len(arr)):
          if s+arr[i] == k:
            c += 1

    return c
```

```python
arr = [1,1,2,3,4,4]
k = 4

numberOfWays(arr, k)
```

```python
arr = [1, 2, 3, 4, 3]
k = 6
numberOfWays(arr, k)
```

# 4. Counting triangles

```python
from collections import defaultdict
def countDistinctTriangles(arr):
    # Write your code here
    hash_map  = defaultdict(list)

    for i in arr:
        hash_map[tuple(sorted(i))].append(i)

    return len(hash_map.keys())


```

```python
arr = [[2, 2, 3], [3, 2, 2], [2, 5, 6]]
countDistinctTriangles(arr)
```

# 5. Encrypted Words

```python
import math

def findEncryptedWord(s):
    # Write your code here
    if not s:
        return ""

    cur_length = len(s)

    if cur_length == 1 or cur_length == 2:
        return s
    if cur_length % 2 == 0:
        mid = cur_length // 2 - 1
    else:
        mid = cur_length // 2
    return s[mid] + findEncryptedWord(s[:mid]) + findEncryptedWord(s[mid+1:])
```
